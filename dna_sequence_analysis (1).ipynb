```python
import pandas as pd


file_name = list(uploaded.keys())[0]

with open(file_name, "r") as file:
    lines = file.readlines()


sequence = ""

for line in lines:
    if not line.startswith(">"):
        sequence += line.strip()


chunk_size = 50

dna_sequences = []

for i in range(0, len(sequence), chunk_size):
    dna_sequences.append(sequence[i:i+chunk_size])


df = pd.DataFrame({
    "Sequence": dna_sequences
})


df.to_csv("dna.csv", index=False)

print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("dna.csv")

df["Length"] = df["Sequence"].apply(len)


df["A_Count"] = df["Sequence"].apply(lambda x: x.count("A"))
df["T_Count"] = df["Sequence"].apply(lambda x: x.count("T"))
df["G_Count"] = df["Sequence"].apply(lambda x: x.count("G"))
df["C_Count"] = df["Sequence"].apply(lambda x: x.count("C"))


df["GC_Content"] = (
    (df["G_Count"] + df["C_Count"]) / df["Length"]
) * 100

df["GC_Content"] = df["GC_Content"].round(2)


print(df.head())


print("\nAverage GC Content:")
print(df["GC_Content"].mean())

print("\nLongest Sequence Length:")
print(df["Length"].max())

print("\nShortest Sequence Length:")
print(df["Length"].min())


high_gc = df[df["GC_Content"] > 60]

print("\nHigh GC Sequences:")
print(high_gc)

sorted_df = df.sort_values(by="GC_Content", ascending=False)

print("\nSorted by GC Content:")
print(sorted_df.head())
TA=df["A_Count"].sum()
TG=df["G_Count"].sum()
TC=df["C_Count"].sum()
TT=df["T_Count"].sum()
nucleotides=["A","T","G","C"]
number_nucleotides=[TA, TT, TG, TC]
plt.bar(nucleotides,number_nucleotides)
plt.xlabel("nucleotides")
plt.ylabel("number of nucleotides")
plt.title("nucleotides in dna ")
plt.show()
labels = ["A", "T", "G", "C"]
sizes = [TA, TT, TG, TC]

plt.figure(figsize=(6,6))

plt.pie(sizes, labels=labels, autopct="%1.1f%%")

plt.title("Nucleotide Composition")

plt.show()
plt.figure(figsize=(6,4))

plt.hist(df["GC_Content"], bins=5)

plt.xlabel("GC Content (%)")
plt.ylabel("Frequency")
plt.title("GC Content Distribution")

plt.show()

```
